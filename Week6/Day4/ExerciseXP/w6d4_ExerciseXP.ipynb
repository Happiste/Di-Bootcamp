##ðŸŒŸ Exercise 1 : Linear Regression Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import pearsonr, spearmanr


house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]
house_sizes = np.array(house_sizes)
house_prices = np.array(house_prices)
plt.scatter(house_sizes, house_prices, color='blue')
plt.xlabel('House Size')
plt.ylabel('House Price')
plt.title('House Size vs. House Price')
plt.show()
```

What is the slope and intercept of the regression line?

```python
slope, intercept = np.polyfit(house_sizes, house_prices, 1)
print(f"Slope: {slope}, Intercept: {intercept.round()}")
```


Predict the price of a house that is 90 square meters.



```python
price_of_90_square_meters = slope * 90 + intercept
print(f"Price of a house with 90 square meters: {price_of_90_square_meters.round()}")
```

Interpret the meaning of the slope in the context of housing prices.


The slope indicates how much the price increases for each additional square meter.It is the price per square meter.

```python

y_pred = slope * house_sizes + intercept

plt.scatter(house_sizes, house_prices, color='blue', label='data')
plt.plot(house_sizes, y_pred, color='red', label='linear regression')

plt.title("Linear regression: Price vs Size")
plt.xlabel("House Size (mÂ²)")
plt.ylabel("House Price")
plt.legend()
plt.grid(True)

plt.show()
```

##ðŸŒŸ Exercise 4 : Correlation Coefficients

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
data.head()
```

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

```python
pearson_corr = pearsonr(data['age'], data['income'])
spearman_corr = spearmanr(data['age'], data['income'])

print(f"Pearson correlation coefficient: {pearson_corr[0]}")
print(f"Spearman correlation coefficient: {spearman_corr[0]}")
```

Insight: Age is strongly related to income in these data.

##ðŸŒŸ Exercise 5 : Linear Relationships
