#Daily Challenge: NumPy, Pandas, and Matplotlib Integration

##1. Data Preparation:

```python
import numpy as np
import pandas as pd

array = np.random.uniform(-5, 35, [10,12])
temp = array.round(1)
temp
```

```python
city_dico = {
    'Paris': 1,
    'Marseille': 2,
    'Lyon': 3,
    'Toulouse': 4,
    'Nice': 5,
    'Nantes': 6,
    'Strasbourg': 7,
    'Montpellier': 8,
    'Bordeaux': 9,
    'Lille': 10
}


df = pd.DataFrame(temp, index=city_dico, columns=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

##2. Data Analysis:

Calculate the annual average temperature for each city.

```python
avg_annual_tmp_by_city = df.mean(axis=1)
avg_annual_tmp_by_city = avg_annual_tmp_by_city.reset_index()
avg_annual_tmp_by_city.columns = ['City', 'Average Temperature']
avg_annual_tmp_by_city
```

Identify the city with the highest and lowest average temperature for the year.

```python
highest_avg_tmp_city = avg_annual_tmp_by_city.loc[avg_annual_tmp_by_city['Average Temperature'].idxmax()]
highest_avg_tmp_city
```

```python
lowest_avg_tmp_city = avg_annual_tmp_by_city.loc[avg_annual_tmp_by_city['Average Temperature'].idxmin()]
lowest_avg_tmp_city
```

##3. Data Visualization:

Global View trend

```python
df.head()
```

```python
df_transposed = df.T.reset_index()
df_transposed.rename(columns={'index': 'Month'}, inplace=True)
df_transposed
```

```python
import plotly.express as px

city = df_transposed.columns[1:]
fig = px.line(df_transposed, x='Month', y=city, title='Global View Trend',
              labels={"value": "Temperature (C)", "variable": "City"})
fig.show()
```

data visualization of the average temperature by city

```python


fig = px.bar(avg_annual_tmp_by_city, x='City', y='Average Temperature', title='Average Temperature by City', color='Average Temperature')
fig.show()
```

A heatmap to have a fast global view

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.xticks(rotation=45)
plt.show()
```

Faceted Histogram of each City to see the occurence of each temperature by city

```python
df_long = df.reset_index().melt(id_vars='index', var_name='Month', value_name='Temperature')
df_long.rename(columns={'index': 'City'}, inplace=True)
df_long
```

```python
from plotnine import *

histogram_plot = (
    ggplot(df_long, aes(x='Month', y='Temperature', fill='Temperature',
                        )) +
    geom_col(alpha=0.8) +
    facet_wrap('~City', ncol=5) +  # Facettage par ville
    theme_minimal() +
    theme(
        figure_size=(14, 8),
        axis_text_x=element_text(rotation=45, ha='right')
    ) +
    labs(
        title="Faceted Histogram: Temperature Distribution by City",
        x="Temperature (C)",
        y="Count"
    )
)

histogram_plot

```
