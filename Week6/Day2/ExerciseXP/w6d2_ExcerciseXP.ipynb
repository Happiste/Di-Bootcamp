##ðŸŒŸ Exercise 6 : Basic Hypothesis *Testing*

```python
# Productivity scores of employees before the training program

np.random.seed(42)
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)
print('The productivity before:\n',productivity_before)
print('The productivity after:\n',productivity_after)
```

```python
import numpy as np
from scipy.stats import ttest_1samp

h0 = 'The mean productivity before and after are the same.'
h1 = 'The mean productivity before and after are different.'

mean_productivity_before = np.mean(productivity_before)
mean_productivity_after = np.mean(productivity_after)
print('The mean productivity before:', mean_productivity_before)
print('The mean productivity after:', mean_productivity_after)
diff_mean = mean_productivity_after - mean_productivity_before
print('The difference between the mean productivity before and after:', diff_mean)
differences = productivity_after - productivity_before
print('The differences between the productivity before and after:\n', differences)
std_dev_diff = np.std(differences, ddof=1)
print('The standard deviation of the differences:', std_dev_diff)
t_statistic, p_value = ttest_1samp(differences, popmean=0)
print('The t-statistic:', t_statistic)
print('The p-value:', p_value)
if p_value < 0.05:
    print('The null hypothesis is rejected.')
else:
    print('The null hypothesis is not rejected')
```

ðŸŒŸ Exercise 8 : Time Series Data Manipulation

```python

# Generate time series dates for the year 2023
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")

# Generate random data for each day
data = np.random.rand(len(dates))

# Create a Pandas Series for the time series
time_series = pd.Series(data, index=dates)

# Display the first few entries
print("Time Series Data:")
print(time_series.head())

# Slicing the time series for specified intervals
jan_to_mar = time_series["2023-01-01":"2023-03-31"]
apr_to_jun = time_series["2023-04-01":"2023-06-30"]
jul_to_sep = time_series["2023-07-01":"2023-09-30"]
oct_to_dec = time_series["2023-10-01":"2023-12-31"]

# Display results
print("\nJanuary to March:")
print(jan_to_mar)
print("\nApril to June:")
print(apr_to_jun)
print("\nJuly to September:")
print(jul_to_sep)
print("\nOctober to December:")
print(oct_to_dec)
```
