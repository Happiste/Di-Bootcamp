ðŸŒŸ Exercise 2: Handling Missing Values

```python
train_data.info()
```

```python
train_data.drop(columns=['Ticket'], inplace=True)
train_data.info()
```

```python
train_data['Cabin'] =train_data['Cabin'].fillna('Unknown')
train_data.info()
```

```python
from sklearn.impute import SimpleImputer
age_imputer = SimpleImputer(strategy='median')
train_data['Age'] = age_imputer.fit_transform(train_data[['Age']])
train_data.info()
```

```python
train_data['Embarked'].unique()
```

```python
train_data['Embarked'].dtype
```

```python
train_data['Embarked'] = train_data['Embarked'].astype(str)

```

```python
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])

train_data.info()
```

ðŸŒŸ Exercise 3: Feature Engineering

```python
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
```

```python
train_data.head()
```

```python
train_data['Title'] = train_data['Name'].str.split(',').str[1].str.split('.').str[0].str.strip()
train_data.head()
```

```python
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
train_data['Embarked'] = LabelEncoder().fit_transform(train_data['Embarked'])
train_data.head()
```

```python
train_data['Fare'].describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
train_data[['Fare_normalize']] = MinMaxScaler().fit_transform(train_data[['Fare']])
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
train_data[['Age_standardize']] = StandardScaler().fit_transform(train_data[['Age']])
```

ðŸŒŸ Exercise 4: Outlier Detection and Handling

```python
train_data['Fare'].describe()
```

```python
Q1 = train_data['Fare'].quantile(0.25)
Q1
```

```python
Q3 = train_data['Fare'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
outlier_condition = (train_data['Fare'] < (Q1 - 1.5 * IQR)) | (train_data['Fare'] > (Q3 + 1.5 * IQR))
outlier_condition.sum()
```

```python

import numpy as np
train_data['Fare_Log'] = train_data['Fare'].apply(lambda x: np.log(x + 1))
```
