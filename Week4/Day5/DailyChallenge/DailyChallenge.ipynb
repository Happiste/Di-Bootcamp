```python
salaries_df[['experience_level', 'salary']].head()
```

```python
grouped_stats = salaries_df.groupby('experience_level')['salary'].agg(['mean', 'median'])
grouped_stats
```

```python
import matplotlib.pyplot as plt

grouped_stats[['mean', 'median']].plot(kind='bar', figsize=(10, 6), title='Mean and Median Salaries by Experience Level')
plt.ylabel('Salary')
plt.show()
```
