```python
#Excercise 1

flag = False
num = 7
if num == 0 or num == 1:
  print(f'{num} is a prime number')
elif num > 1:
  for i in range(2, num):
    if (num % i) == 0:
      flag = True

if flag:
  print(f'{num} is not a prime number')
else:
  print(f'{num} is a prime number')
```
