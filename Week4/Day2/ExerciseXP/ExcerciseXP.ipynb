```python
for column in time.columns:
    dtype = time[column].dtype
    if pd.api.types.is_numeric_dtype(dtype):
        print(f"{column}: Quantitative (Reason: Valeurs numériques représentant des mesures ou des comptes)")
    else:
        print(f"{column}: Qualitative (Reason: Données catégoriques ou descriptives)")
```
