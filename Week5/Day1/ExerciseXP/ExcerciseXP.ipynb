ðŸŒŸ Exercise 2 : Creating a Line Plot for Temperature Variation

```python
import random as r
import matplotlib.pyplot as plt
temp = [r.randint(20, 40) for x in range(7)]
day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# day = [x for x in range(len(temp))]
```

```python
plt.figure()
plt.plot(day, temp, label='temperature')
plt.xlabel('Days')
plt.ylabel('Temperature in Celsius')
plt.title('Temperature Variation Over Days')
plt.legend()
plt.show()
```

ðŸŒŸ Exercise 3 : Visualizing Monthly Sales with a Bar Chart

```python
sales = [r.randint(1000, 5000) for x in range(12)]
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.figure(figsize=(14, 6))
plt.bar(month, sales, label='sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
avg_revenue = df['revenue'].mean()
avg_revenue
```

```python
df.head()
```

```python
df['date']= pd.to_datetime(df['date'], errors='coerce')
df.head()
```

```python
df['quarter'] = df['date'].dt.quarter
df.head()
```

```python
quarter_revenue = df.groupby('quarter')['revenue'].sum()
quarter_revenue
```

```python
plt.figure(figsize=(12, 6))
plt.bar(quarter_revenue.index, quarter_revenue.values, label='revenue')
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.title('Quarterly Revenue')
plt.legend()
plt.show()
```
