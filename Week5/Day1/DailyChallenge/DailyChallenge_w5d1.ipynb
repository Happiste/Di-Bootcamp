```python
import pandas as pd
happiness_data = pd.read_csv('2019.csv')
happiness_data.head()
```

```python
happiness_data.head(15)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(happiness_data['Social support'], happiness_data['Score'], alpha=0.7, edgecolor='k')
plt.title('Social Support vs. Happiness Score', fontsize=16)
plt.xlabel('Social Support', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


happiness_data['Region'] = pd.cut(
    happiness_data['GDP per capita'], bins=4, labels=['Low', 'Medium', 'High', 'Very High']
)

region_data = happiness_data.groupby('Region')[['GDP per capita', 'Healthy life expectancy']].mean()

fig, ax = plt.subplots(figsize=(10, 6))
region_data['GDP per capita'].plot(kind='bar', ax=ax, color='lightblue', alpha=0.8, width=0.4, position=1, label='GDP per Capita')
region_data['Healthy life expectancy'].plot(kind='line', ax=ax, color='orange', marker='o', label='Healthy Life Expectancy')
ax.set_title('Regional Comparison of GDP and Healthy Life Expectancy', fontsize=16)
ax.set_xlabel('Region', fontsize=12)
ax.set_ylabel('Values', fontsize=12)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
