```python
import pandas as pd
from plotnine import *

df = pd.read_csv('Iris.csv')
df.head()
```

```python
scatter_plot = (
    ggplot(df, aes(x='SepalLengthCm', y='SepalWidthCm', color='Species')) +
    geom_point() +
    labs(x='Sepal Length (cm)', y='Sepal Width (cm)')
)
scatter_plot.show()
```

```python
import plotly.express as px
import pandas as pd


df = pd.read_csv('Iris.csv')


fig = px.scatter(
    df,
    x='SepalLengthCm',
    y='SepalWidthCm',
    color='Species',
    marginal_x='histogram',
    marginal_y='histogram',
    title="Scatter Plot with Marginal Histograms"
)

fig.update_layout(
    width=800,
    height=600
)
fig.show()
```

Exercise 2: Interactive Facet Grid with Conditional Elements

```python
import pandas as pd

# Charger le jeu de données depuis le dépôt GitHub de Seaborn
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)
facet_plot = (
    ggplot(df, aes(x='horsepower', y='mpg', color='origin'))
    + geom_point()
    + facet_wrap('~origin')
    + labs(title="Scatter Plot with Facet Wrapping",
         x="Horsepower",
         y="Miles Per Gallon",
         )
    + theme(figure_size=(12, 8))
    + theme_minimal()
)

facet_plot.show()
```
