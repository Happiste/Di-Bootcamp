```python
# Re-import necessary libraries and redefine the data after the state reset

import pandas as pd

data = {
    'Book Title': ['The Great Gatsby', 'To Kill a Mockingbird', '1984', 'Pride and Prejudice', 'The Catcher in the Rye','1985'],
    'Author': ['F. Scott Fitzgerald', 'Harper Lee', 'George Orwell', 'Jane Austen', 'J.D. Salinger','Harper Lee'],
    'Genre': ['Classic', 'Classic', 'Dystopian', 'Classic', 'Classic','Classic'],
    'Price': [10.99, 8.99, 7.99, 11.99, 9.99, 6.7],
    'Copies Sold': [500, 600, 800, 300, 450, 1000]
}

df = pd.DataFrame(data)

# Step 2: Exploring the Dataset

```

```python

# View the first few rows of the DataFrame
head_data = df.head()
head_data
```

```python
# Get a statistical summary of the numerical columns
stats_summary = df.describe()
stats_summary
```

```python


# Sort the DataFrame by Price (ascending) and by Copies Sold (descending)
sorted_by_price = df.sort_values(by="Price", ascending=True)
sorted_by_copies = df.sort_values(by="Copies Sold", ascending=False)
print("sorted by price",sorted_by_price)
```

```python
print("sorted by copies",sorted_by_copies)
```

```python

# Filter books with Price above 9.00 or Genre as 'Classic'
filtered_by_price = df[df["Price"] > 9.00]
filtered_by_price

```

```python
filtered_by_genre = df[df["Genre"] == "Classic"]
filtered_by_genre
```

```python

# Group by Author and sum up Copies Sold
grouped_by_author = df.groupby("Author")["Copies Sold"].sum()
grouped_by_author
```

## Change the prices of every author whose name is 'Harper Lee' and the price is below 10 to 8


```python
#how to do with a loop
for index in range(len(df)):
    if df["Author"][index] == "Harper Lee" and df["Price"][index] < 10:
        df["Price"][index] = 8
```

```python
df
```

```python
#hpow to do with a loc (we will learn this later in the week)
df.loc[(df["Author"] == "Harper Lee") & (df["Price"] < 10), "Price"] = 8
```
