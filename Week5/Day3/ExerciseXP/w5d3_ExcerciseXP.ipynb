
Exercise 1: Creating a Scatter Plot with Aesthetic Mappings


```python
import pandas as pd

df = pd.read_csv('Iris.csv')
print(df.head())
print(df.info())
print(df.describe())
```

```python
from plotnine import *

options.figure_size = (10, 5)

scatter_plot = (
    ggplot(df, aes(x='SepalLengthCm', y='SepalWidthCm',color='Species'))
    + geom_point()
    + labs(title="Scatter Plot with Custom Colors",
             x="Sepal Length in cm",
             y="Sepal Width in cm",
             )
)

scatter_plot.show()
```

Exercise 2: Building a Stacked Bar Chart

```python
df['Region'] = df['Species'].map({
    'Iris-setosa': 'Jerusalem',
    'Iris-versicolor': 'Neguev',
    'Iris-virginica': 'Mitzpe Ramon'
})
df.head()
```

```python
bar_plot = (
    ggplot(df, aes(x='Species', fill='Region'))
    + geom_bar()
    + labs(title="Stacked Bar Chart",
           x="Species",
           y="Count",
           )
)

bar_plot.show()
```

Exercise 3: Exploring Facet Wrapping in Plotnine

```python
import pandas as pd

# Charger le jeu de données depuis le dépôt GitHub de Seaborn
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

# Afficher les premières lignes du DataFrame
print(df.head())
print(df.info())
```

```python
df = df.dropna()
df.info()
```

```python
facet_plot = (
    ggplot(df, aes(x='horsepower', y='mpg', color='origin'))
    + geom_point()
    + facet_wrap('~origin')
    + labs(title="Scatter Plot with Facet Wrapping",
         x="Horsepower",
         y="Miles Per Gallon",
         )
    + theme(figure_size=(12, 8))
    + theme_minimal()
)

facet_plot.show()
```

Exercise 6: Interactive Heatmap using Flight Data

```python
df = pd.read_csv('Airlines.csv')
df.head()
```

```python
df.info()
```

```python
import plotly.express as px

flight_counts = df.groupby(['AirportFrom', 'AirportTo']).size().reset_index(name='FlightFrequency')
flight_counts.head()
```

```python

pivot_table = flight_counts.pivot_table(
    index='AirportTo',
    columns='AirportFrom',
    values='FlightFrequency',
    aggfunc='sum',
)

print(pivot_table)
```

```python
import plotly.graph_objects as go

# Créer la heatmap
heatmap = go.Figure(data=go.Heatmap(
    z=pivot_table.values,  # Valeurs pour la couleur
    x=pivot_table.columns,  # Villes de départ
    y=pivot_table.index,  # Villes d'arrivée
    colorscale='Viridis',  # Palette de couleurs
    hoverongaps=False  # Gérer les données manquantes
))

# Ajouter des titres et des labels
heatmap.update_layout(
    title="Interactive Heatmap of Flight Frequencies",
    xaxis_title="Departure Cities",
    yaxis_title="Arrival Cities"
)

# Afficher la heatmap
heatmap.show()
```

Exercise 7: Visualizing Large Time-Series Data with Matplotlib

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Définir le ticker pour l'indice S&P 500
ticker = "^GSPC"

# Télécharger les données des 10 dernières années
data = yf.download(ticker, start="2013-01-01", end="2023-12-31")

# Afficher un aperçu des données

data.head()

```

```python
data.to_csv('S&P500.csv')
df = pd.read_csv('S&P500.csv')
df.head()
```

```python
df = df.dropna()
df.drop(index=df.index[0], axis=0, inplace=True)
df.rename(columns={'Price': 'Date'}, inplace=True)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
numeric_columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']


for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')


print(df.dtypes)
print(df.isna().sum())
print(df.shape)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='S&P 500 Closing Price')
plt.title("S&P 500 Closing Prices (2013-2023)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Closing Price (USD)", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(x=df['Date'], y=df['Close'])])
fig.update_layout(
    title="S&P 500 Closing Prices (2013-2023)",
    xaxis_title="Date",
    yaxis_title="Closing Price (USD)",
)
fig.show()
```
