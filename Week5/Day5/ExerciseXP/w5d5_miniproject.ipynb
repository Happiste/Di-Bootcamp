```python
superstore['Order Date'] = pd.to_datetime(superstore['Order Date'], format='%d/%m/%Y')
superstore['Ship Date'] = pd.to_datetime(superstore['Ship Date'], format='%d/%m/%Y')
```

```python
superstore.info()
```

```python
superstore['Sales'] = superstore['Sales'].str.replace(',', '.').astype(float)
superstore['Profit'] = superstore['Profit'].str.replace(',', '.').astype(float)
superstore.info()
```

```python
grouped_sales_by_state = superstore.groupby('State')['Sales'].sum()
```

###What are the Top 20 cities by Sales ? What about the Top 20 cities by Profit ? Are there any difference among cities in profitability ? (Identify the top 20 cities based on total sales and total profit and analyze differences in profitability among these cities.)



```python
grouped_cities = superstore.groupby('City').agg({'Sales': 'sum', 'Profit': 'sum'})
top_20_cities_sales = grouped_cities.sort_values(by='Sales', ascending=False).head(20)
top_20_cities_profit = grouped_cities.sort_values(by='Profit', ascending=False).head(20)
top_20_cities_sales = top_20_cities_sales.reset_index()
top_20_cities_profit = top_20_cities_profit.reset_index()
top_20_merge = top_20_cities_profit.merge(top_20_cities_sales, how='outer', suffixes=('_profit', '_sales'))
top_20_merge['Profitability (%)'] = (top_20_merge['Profit'] / top_20_merge['Sales']) * 100
top_20_merge.sort_values(by='Profitability (%)', ascending=False)

```

```python
grouped_customers = superstore.groupby('Customer Name').agg({'Sales': 'sum'})
top_20_customers_sales = grouped_customers.sort_values(by='Sales', ascending=False).head(20)
top_20_customers_sales = top_20_customers_sales.reset_index()
top_20_customers_sales
```

```python
top_20_customers_sales['Cumulative Sales'] = top_20_customers_sales['Sales'].cumsum()
total_sales = top_20_customers_sales['Sales'].sum()
top_20_customers_sales['Cumulative Percentage'] = (top_20_customers_sales['Cumulative Sales'] / total_sales) * 100
```

```python
import matplotlib.pyplot as plt

grouped_customers = superstore.groupby('Customer Name').agg({'Sales': 'sum'})


top_20_customers_sales = grouped_customers.sort_values(by='Sales', ascending=False).head(20)


top_20_customers_sales['Cumulative Sales'] = top_20_customers_sales['Sales'].cumsum()
total_sales = top_20_customers_sales['Sales'].sum()
top_20_customers_sales['Cumulative Percentage'] = (top_20_customers_sales['Cumulative Sales'] / total_sales) * 100

fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.bar(top_20_customers_sales.index, top_20_customers_sales['Sales'], color='skyblue', label='Sales')
ax1.set_xlabel('Customer Name', fontsize=12)
ax1.set_ylabel('Sales', fontsize=12)
ax1.set_title('Cumulative Sales Curve for Top 20 Customers', fontsize=16)
ax1.tick_params(axis='x', rotation=90)


ax2 = ax1.twinx()
ax2.plot(top_20_customers_sales.index, top_20_customers_sales['Cumulative Percentage'], color='orange', marker='o', label='Cumulative Sales (%)')
ax2.axhline(y=80, color='red', linestyle='--', label='80% Threshold')
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=12)


fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))


plt.tight_layout()
plt.show()



```

* Cities or states with high sales but low or negative profit indicate inefficient operations, high costs, or discounts. > Maybe revising pricing or cost structures.

* Cities with High-Sales, High-Profit should be prioritized for investment and growth

* Cities with Low-Sales, High-Profit. These locations are profitable but underutilized maybe increase awareness through marketing campaigns.
