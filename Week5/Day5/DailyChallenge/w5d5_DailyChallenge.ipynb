## 1. Data Preparation: DataCleaning & processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

superstore = pd.read_csv('superstore.csv')
superstore.head()
```

```python
superstore.info()
```

```python
superstore.drop(columns=['记录数'], inplace=True)
superstore.info()
```

```python
superstore['Order.Date'] = pd.to_datetime(superstore['Order.Date'], format='%Y-%m-%d %H:%M:%S.%f')
superstore['Ship.Date'] = pd.to_datetime(superstore['Ship.Date'], format='%Y-%m-%d %H:%M:%S.%f')
superstore.info()
```

```python
superstore.head(15)
```

```python
Sales_over_years = superstore.groupby(superstore['Order.Date'].dt.year)['Sales'].sum()
Sales_over_years = Sales_over_years.reset_index()
Sales_over_years
```

```python
Profit_over_years = superstore.groupby(superstore['Order.Date'].dt.year)['Profit'].sum()
Profit_over_years = Profit_over_years.reset_index()
Profit_over_years
```

##2. Data Visualization with Plotly:

```python
import plotly.express as px
from plotly.subplots import make_subplots as ms

fig = ms(rows=2, cols=1, subplot_titles=["Sales over the Years", "Profit over the Years"])
fig.add_trace(px.line(Sales_over_years, x='Order.Date', y='Sales').data[0], row=1, col=1)
fig.add_trace(px.line(Profit_over_years, x='Order.Date', y='Profit').data[0], row=2, col=1)
fig.update_layout(template='plotly_dark')
fig.update_xaxes(title_text="Year", row=1, col=1)
fig.update_yaxes(title_text="Sales ($)", row=1, col=1)
fig.update_xaxes(title_text="Year", row=2, col=1)
fig.update_yaxes(title_text="Profit ($)", row=2, col=1)



fig.show()
```

Build an interactive map to visualize sales distribution by country.

```python
Sales_by_country = superstore.groupby('Country')['Sales'].sum()
Sales_by_country = Sales_by_country.reset_index()
Sales_by_country.sort_values(by='Sales', ascending=True, inplace=True)
Sales_by_country
```

```python
Sales_by_country['Country'].unique()
```

```python
fig = px.choropleth(Sales_by_country, locations='Country',  locationmode='country names', color='Sales', hover_name='Country',
                    projection='natural earth', title='Sales Distribution by Country')
fig.show()
```

##3. Data Visualization with Plotnine:

```python
top_product_by_sales = superstore.groupby('Product.Name')['Sales'].sum()
top_product_by_sales = top_product_by_sales.reset_index()
top_10_product_by_sales = top_product_by_sales.nlargest(10, 'Sales')
top_10_product_by_sales.reset_index(drop=True, inplace=True)
top_10_product_by_sales
```

```python
from plotnine import *

bar_plot = (
    ggplot(top_10_product_by_sales, aes(x='Product.Name', y='Sales', fill='Sales'))
    + geom_bar(stat='identity')
    + labs(title='top 10 Product',
           x='Product Name',
           y='Sales')
    + theme_classic()
    + theme(figure_size=(12, 6), axis_text_x=element_text(rotation=45)))


bar_plot.show()


```

```python
scatter_plot = (
    ggplot(superstore, aes(x='Discount', y='Profit', color='Profit'))
    + geom_point(alpha=0.7)
    + geom_smooth(method='lm')

    + labs(title='Discount vs Profit',
           x='Discount',
           y='Profit'
         )
    + theme_grey()
    + theme(figure_size=(10, 6))
)
scatter_plot.show()
```
